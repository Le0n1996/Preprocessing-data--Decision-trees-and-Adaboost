Вот более подробное описание признаков:

 - school - школа, в которой учится студент (бинарный) ("GP" - Gabriel Pereira или "MS" - Mousinho da Silveira)

 - address - откуда студент (бинарный) ("U" - urban или "R" - rural)
 - famsize - размер семьи (бинарный) ("LE3" - меньше или равно 3 или "GT3" - больше 3)
 - Pstatus - в каких отношениях родители (бинарный) ("T" - живут вместе "A" - раздельно)
 - Medu - образование матери (числовой) (0 - никакого,  1 - начальное образование (4 класса), 2 – от 5 до 9 классов, 3 – среднеспециальное или 4 – высшее)
 - Fedu - образование отца (числовой) (0 - никакого,  1 - начальное образование (4 класса), 2 – от 5 до 9 классов, 3 – среднеспециальное или 4 – высшее)
 - Mjob - работа матери (номинальный) ("teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other")
 - Fjob - работа отца (номинальный) ("teacher", "health" care related, civil "services" (e.g. administrative or police), "at_home" or "other")
 - reason - причина выбора школы (номинальный) (близко к дому — "home", репутация школы — "reputation", "course" предпочтение некоторым предметам или "other")
 - guardian - опекун (номинальный) ("mother", "father" или "other")
 - traveltime - время от дома до школы (числовой) (1 - меньше 15 мин., 2 - 15 до 30 мин., 3 - 30 мин. до 1 часа, или 4 - больше 1 часа)
 - studytime - количество часов обучения в неделю (числовой) (1 - меньше 2 часов, 2 - от 2 до 5 часов, 3 - от 5 до 10 часов, или 4 - больше 10 часов)
 - failures - количество ранее не сданных предметов (числовой) (n если 1 <= n < 3, иначе 4)
 - schoolsup - дополнительные занятия (бинарный) (yes or no)
 - famsup - помощь от семьи при выполнении заданий (бинарный) (yes or no)
 - paid - дополнительные платные занятия (бинарный) (yes or no)
 - activities - внеклассная деятельность (бинарный) (yes or no)
 - nursery - посещал детский сад (бинарный) (yes or no)
 - higher - желание высшего образования (бинарный) (yes or no)
 - internet - домашний интернет (бинарный) (yes or no)
 - romantic - состоит в романтических отношениях (бинарный) (yes or no)
 - famrel - насколько хорошо отношения в семье (числовой) (от 1 - очень плохие до 5 - превосходные)
 - freetime - наличие свободного времени после школы (числовой) (от 1 - очень мало до 5 - очень много)
 - goout - гуляет с друзьями (числовой) (от 1 - редко до 5 - очень часто)
 - Dalc - употребление алкоголя в будние дни (числовой) (от 1 - очень редко до 5 - очень часто)
 - Walc - употребление алкоголя в выходные (числовой) (от 1 - очень редко до 5 - очень часто)
 - health - текущее состояние здоровья (числовой) (от 1 - очень плохое до 5 - очень хорошее)
 - absences - количество школьных пропусков (числовой) (от 0 до 93)
 - G1 - оценка за первый семестр (числовой) (от 0 до 20)
 - G2 - оценка за второй семестр (числовой) (от 0 до 20)
 - G3 - итоговая оценка (числовой) (от 0 до 20)
 - alc_week - употребление алкоголя за неделю (числовой) (от 1 - очень редко до 5 - очень часто)
 - Alc - является ли студент алкоголиком (бинарный) (0 или 1)

Статья утверждает, что мы можем не рассматривать признаки famsize, Fedu, Fjob, reason, guardian, famrel; признаки Dalc и Walc мы преобразуем вначале в единый признак alc, а затем, так как мы хотим решать задачу классификации, в признак Alc:

```python
students = pd.concat([st_mat, st_por], ignore_index=True)
# как сказано в статье, мы можем напрямую объединить два датасета
# (несмотря на то, что признаки G1, G2, G3 здесь обозначают, вообще говоря, разные вещи)
students = students.drop(['famsize', 'Fedu', 'Fjob', 'reason', 'guardian', 'famrel'], axis='columns')
students[-10:]
```

```python
alc_week = (students.Walc*2 + students.Dalc*5)/7 # как в статье
print(alc_week[-10:])
Alc = pd.DataFrame({'Alc': (alc_week >= 3).astype(int)}) # как в статье - если меньше 3, то еще не алкоголик
Alc[-10:]
```

```python
students = students.drop(['Walc', 'Dalc'], axis=1).join(Alc)
students[:10]
```

```python
no_space = students.dropna(axis='index')
print(students.index.size - no_space.index.size) # сколько школьников оставили пропуски
```

Пропусков в данных вообще нет.

Так как мы, руководствуясь статьей, удалили признаки Fjob, reason, guardian, то у нас остается только один номинальный признак - Mjob.

```python
print(students.Mjob.unique()) # как видим, принимает 5 значений - at_home, health, other, services, teacher.
```

```python
Mjob_bin = pd.get_dummies(students.Mjob)
Mjob_bin[-10:]
```

```python
students = students.rename(columns={'health':'st_health'}) # иначе неразбериха
```

```python
students = pd.concat([students.drop(['Mjob'], axis=1), Mjob_bin], axis='index')
students[-10:]
```

```python
binary = [i for i in students.columns if students[i].dtype.name == 'object']
print(binary) # все бинарные признаки, которые мы будем кодировать 0 и 1
```

```python
description = students[binary].describe()
frequent = description[description.index == 'top'] # Это наиболее частое значение для каждого признака
frequent
```

Будем для каждого признака кодировать то значение, которое встречается чаще, единицей, а то, которое встречается реже - нулем.

```python
for i in binary:
    frequent_value = description[i]['top']
    students.loc[students[i] != frequent_value, i] = 0
    students.loc[students[i] == frequent_value, i] = 1
students[-15:]
```

Теперь все признаки числовые, можем начинать классификацию.
